# Easy to select

I want to look at the C peptide area under curve, during the OGTT of my patients.

**For this i need to:**
    
    - Select only the patient that were not excluded from the study
    - Select only the patient that didn't change medication that could affect glucose metabolism
    - Select only patient that performed an OGTT
    - Select only patient that i have information for M0 and M3
    - Discriminate between Inuline and Maltodextrine 

```python
f4g_select = f4g.loc[f4g['Exclusion'] == 'No']
f4g_select = f4g.loc[f4g['Change of Insuline medication'] == 'NO']
f4g_select = f4g.loc[f4g['OGTT C peptide'] == 'Yes']
```

I have already written some bit of code to get paired values, and graph them but also, perform statistics on the distribution

# It makes making relation between datas very easy.

Let's look with what Fasting C peptide is correlated to. 

```python
corr_matrix = f4g_start.corr('pearson')

corr_matrix['Average Fasting C peptide'].sort_values(ascending=False)
```

# Wow, we have some cool correlations here, i wonder what they look like.

```python
from pandas.plotting import scatter_matrix

attributes = ['Average Fasting C peptide','Fibroscan CAP median', "gGT", 
             'HOMA ISI', 'Average Fasting Insulin']

scatter_matrix(f4g_start[attributes], figsize=(13,10))
plt.show()
```

# We can do crazy stuff pretty fast.

```python
from analysis_script import spearman_p_value_DataFrame, correct_p_values, do_heatmap, revert_map

# Thoses are function i wrote to get spearman correlation and their p-value. Since we do so many comparisons, we have
# to correct thoses p-values with some method( i choosed holm, because it's the most common used to correct correlation
# tables)
            
spearman_p_value, corrspearman = spearman_p_value_DataFrame(f4g_start[attributes])
corelation_corrected = correct_p_values(spearman_p_value)
```

```python
revert_map(corelation_corrected)
do_heatmap(corrspearman)
```
